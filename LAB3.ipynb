```python
import zipfile
import pandas as pd
from IPython.display import display  

zip_path = 'C:/Users/yaros/Downloads/Lab3_Dataset.zip'
extracted_folder = 'C:/Users/yaros/Downloads/'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_folder)

scimagojr_path = extracted_folder + 'scimagojr.xlsx'

df_scimagojr = pd.read_excel(scimagojr_path)

top_countries = df_scimagojr[df_scimagojr['Rank'] <= 15]

df_scimagojr['Country'] = df_scimagojr['Country'].replace({
    "United States of America": "United States"
})

df_energy['Country'] = df_energy['Country'].replace({
    "United States of America": "United States"
})

df_gdp['Country Name'] = df_gdp['Country Name'].replace({
    "United States of America": "United States"
})

df_energy['Country'] = df_energy['Country'].replace({
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom"
})

df_gdp['Country Name'] = df_gdp['Country Name'].replace({
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom"
})


final_data = top_countries.merge(df_energy, on='Country', how='inner')

gdp_years = [str(year) for year in range(2006, 2016)]

gdp_filtered = df_gdp[['Country Name'] + gdp_years]

final_data = final_data.merge(gdp_filtered, left_on='Country', right_on='Country Name', how='inner')

final_data = final_data[['Country', 'Rank', 'Documents', 'Citable documents', 'Citations', 
                         'Self-citations', 'Citations per document', 'H index', 
                         'Energy Supply', 'Energy Supply per Capita', '% Renewable'] + gdp_years]

final_data.set_index('Country', inplace=True)

display(final_data)
```

```python
def average_gdp(final_data):
    gdp_columns = [str(year) for year in range(2006, 2016)]
    
    final_data['Average GDP'] = final_data[gdp_columns].mean(axis=1)
    
    return final_data['Average GDP'].sort_values(ascending=False)

average_gdp_series = average_gdp(final_data)

display(average_gdp_series)
```

```python
def gdp_change(final_data):
    # Обчислення середнього ВВП та отримання країни з 5-м середнім ВВП
    average_gdp_series = final_data['Average GDP'].sort_values(ascending=False)
    
    # Назва країни з 5-м середнім ВВП
    country_fifth = average_gdp_series.index[4]  # Індекси починаються з 0, тому 4 - це 5-й
    
    # Обчислення зміни ВВП за останні 10 років
    gdp_last_year = final_data.loc[country_fifth, '2015']
    gdp_first_year = final_data.loc[country_fifth, '2006']
    
    gdp_change_value = gdp_last_year - gdp_first_year
    
    # Повернення кортежу з назвою країни та зміною ВВП
    return (country_fifth, gdp_change_value)

# Виклик функції та збереження результату
gdp_change_result = gdp_change(final_data)

# Перегляд результату
print(gdp_change_result)
```

```python
def max_renewable_energy(final_data):
    # Знаходження максимального відсотка відновлюваних джерел енергії
    max_renewable_index = final_data['% Renewable'].idxmax()
    max_renewable_value = final_data.loc[max_renewable_index, '% Renewable']
    
    # Повернення кортежу з назвою країни та відсотком
    return (max_renewable_index, max_renewable_value)

# Виклик функції та збереження результату
max_renewable_result = max_renewable_energy(final_data)

# Перегляд результату
print(max_renewable_result)
```

```python
def population_estimation(final_data):
    # Розрахунок населення
    final_data['Population'] = final_data['Energy Supply'] / final_data['Energy Supply per Capita']
    
    # Сортування за оцінкою населення у спадному порядку
    sorted_population = final_data.sort_values(by='Population', ascending=False)

    # Отримання шостої країни за оцінкою населення
    sixth_country = sorted_population.iloc[5]  # Індекс 5 для шостого місця
    country_name = sixth_country.name  # Назва країни
    population = sixth_country['Population']  # Оцінка населення

    return (country_name, population)

# Виклик функції та збереження результату
sixth_country_result = population_estimation(final_data)

# Перегляд результату
print(sixth_country_result)
```

```python
def estimate_population_and_correlation(final_data):
    # Оцінка населення: Energy Supply / Energy Supply per Capita
    final_data['Population'] = final_data['Energy Supply'] / final_data['Energy Supply per Capita']
    
    # Оцінка кількості цитованих документів на одну особу
    final_data['Citations per Person'] = final_data['Citations'] / final_data['Population']
    
    # Розрахунок кореляції Пірсона між 'Citations per Person' і 'Energy Supply per Capita'
    correlation = final_data['Citations per Person'].corr(final_data['Energy Supply per Capita'])

    return correlation

# Виклик функції та збереження результату
correlation_result = estimate_population_and_correlation(final_data)

# Перегляд результату
print(correlation_result)
```

```python
def renewable_energy_median_indicator(final_data):
    # Обчислити медіану для стовпця '% Renewable'
    median_renewable = final_data['% Renewable'].median()

    
    final_data['Renewable Indicator'] = (final_data['% Renewable'] >= median_renewable).astype(int)

    
    return final_data['Renewable Indicator'].sort_values(ascending=True)

renewable_indicator_result = renewable_energy_median_indicator(final_data)
print(renewable_indicator_result)
```

```python
import pandas as pd
from IPython.display import display

def group_by_continent(final_data):
    
    ContinentDict = {
        'China': 'Asia',
        'United States': 'North America',
        'Japan': 'Asia',
        'United Kingdom': 'Europe',
        'Russian Federation': 'Europe',
        'Canada': 'North America',
        'Germany': 'Europe',
        'India': 'Asia',
        'France': 'Europe',
        'South Korea': 'Asia',
        'Italy': 'Europe',
        'Spain': 'Europe',
        'Iran': 'Asia',
        'Australia': 'Australia',
        'Brazil': 'South America'
    }

    # Додати новий стовпець 'Continent' в final_data
    final_data['Continent'] = final_data.index.map(ContinentDict)

    # Розрахунок оцінки населення
    final_data['Estimated Population'] = final_data['Energy Supply'] / final_data['Energy Supply per Capita']

    # Групування за континентами
    grouped = final_data.groupby('Continent')['Estimated Population'].agg(
        size='count',
        sum='sum',
        mean='mean',
        std='std'
    )

    # Повернення результату як DataFrame
    return grouped

# Виклик функції та збереження результату
continent_stats = group_by_continent(final_data)

# Вивід результату у красивій таблиці
display(continent_stats)
```
